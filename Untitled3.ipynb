Σε ένα df κρατάμε μόνο τις τιμες των date & text

Μετατρέπουμε το date σε datetime

```python
ert_articles_df.loc[:, 'date'] = pd.to_datetime(ert_articles_df['date'], format='%Y-%m-%d')
ert_articles_df.dtypes
```

Αφαιρούμε ό,τι url μπορεί να υπάρχει στα κείμενα των άρθρων.

```python
ert_articles_df.loc[:, 'text'] = ert_articles_df['text'].str.replace(r'https?:\/\/.*[\r\n]*'," ", regex=True)
```

Φτιάχνουμε ένα vectorizer με τις συναισθηματικά φορτισμένες λέξεις



```python

vec = CountVectorizer(analyzer = 'word', vocabulary = emolex_df.word,
                      lowercase=False,
                      strip_accents = 'unicode',
                      stop_words= list(nlp.Defaults.stop_words),
                      ngram_range=(1, 2))
```

Bλέπουμε πόσες φορές η κάθε συναισθηματικά φορτισμένη λέξη αναφέρεται στα άρθρα που μελετάμε.

```python



matrix = vec.fit_transform(ert_articles_df['text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
wordcount_df.shape
```

Με την παρακάτω εντολή ελέγχουμε μόνο τις συναισθηματικά φορτισμένες λέξεις

```python

positive_words = emolex_df[emolex_df.Positive == 1]['word']

negative_words = emolex_df[emolex_df.Negative == 1]['word']

angry_words = emolex_df[emolex_df.Anger == 1]['word']

anticipation_words = emolex_df[emolex_df.Anticipation == 1]['word']

disgust_words = emolex_df[emolex_df.Disgust == 1]['word']

fear_words = emolex_df[emolex_df.Fear == 1]['word']

joy_words = emolex_df[emolex_df.Joy == 1]['word']

sadness_words = emolex_df[emolex_df.Sadness == 1]['word']

surprise_words = emolex_df[emolex_df.Surprise == 1]['word']

trust_words = emolex_df[emolex_df.Trust == 1]['word']
```

στήλες συναισθημάτων

```python
ert_articles_df['anger'] = wordcount_df[angry_words].sum(axis=1)
ert_articles_df['positive'] = wordcount_df[positive_words].sum(axis=1)
ert_articles_df['joy'] = wordcount_df[joy_words].sum(axis=1)
ert_articles_df['disgust'] = wordcount_df[disgust_words].sum(axis=1)
ert_articles_df['surprise'] = wordcount_df[surprise_words].sum(axis=1)
ert_articles_df['trust'] = wordcount_df[trust_words].sum(axis=1)
ert_articles_df['anticipation'] = wordcount_df[anticipation_words].sum(axis=1)
ert_articles_df['sadness'] = wordcount_df[sadness_words].sum(axis=1)
ert_articles_df['negative'] = wordcount_df[negative_words].sum(axis=1)
ert_articles_df['fear'] = wordcount_df[fear_words].sum(axis=1)
```
